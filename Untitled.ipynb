```python
from sklearn.preprocessing import LabelEncoder
```

```python
label = LabelEncoder()
```

```python
df_train.info()
```

```python
df_train['Stage_fear'] = label.fit_transform(df_train['Stage_fear'])
df_test['Stage_fear'] = label.transform(df_test['Stage_fear'])
```

```python
df_train['Drained_after_socializing'] = label.fit_transform(df_train['Drained_after_socializing'])
df_test['Drained_after_socializing'] = label.transform(df_test['Drained_after_socializing'])
```

```python
df_train['Personality'] = label.fit_transform(df_train['Personality'])
```

```python
df_train.info()
```

```python
from xgboost import XGBClassifier
import xgboost as xgb
```

```python
model = XGBClassifier(
    n_estimators=500,        
    learning_rate=0.05,      
    max_depth=6,             
    subsample=0.8,           
    colsample_bytree=0.8,    
    gamma=1,                
    reg_alpha=0.1,           
    reg_lambda=1,            
    use_label_encoder=False, 
    eval_metric='logloss',   
    random_state=42,
    n_jobs=-1                
)
```

```python
df_train = df_train.drop('id',axis = 1)
df_test = df_test.drop('id',axis = 1)
```

```python
X_train = df_train.drop('Personality',axis = 1)
y_train = df_train['Personality']
X_test = df_test.copy()
```

```python
model.fit(X_train,y_train)
```

```python
import xgboost as xgb
xgb.plot_importance(model, max_num_features=10); plt.show()
```

```python
y_pred = model.predict(X_test)
```

```python
sub = pd.read_csv('sample_submission.csv')
```

```python
sub["Personality"] = sub["Personality"].map({1: "Extrovert", 0: "Introvert"})
```

```python
sub['Personality'] = y_pred
```
